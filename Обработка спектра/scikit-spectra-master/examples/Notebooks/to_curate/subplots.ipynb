## Slicing into 11 subsets (OrderedDict)

```python
def slice(ts,l):  
## ts is the timespectra to slice. 
## l is the number of curves in each subset.
## Here we slice ts according to its columns, which means we slice timepoints to different sets.

    subsets = [] #build a list of pairs to save subset name and subset dataframe
    for i in range(0,len(ts.columns),l): #slice the timespectra into 11 sub datasets
        #print i
        subsets.append( ('ts'+str(i/l) , ts.ix[:,i:i+l-1]) )
    subsets = OrderedDict(subsets)
    return subsets
```

```python
subts = slice(ts,9)  
# subts is an ordered dictionary which saves the sub datasets of ts. 
# Here we slice ts into 11 sets. Each of them has 9 curves.
subts.keys()
```

```python
def subplot(subsets): 
## The function do subplots, which requires passing the dictionary of subsets into.

    pylab.rcParams['figure.figsize'] = 16,16
    n = len(subsets)
    f, sub = plt.subplots(n/3+1,n%3+1)
    for i in range(0,n):
        sub[i/3,i%3].plot(ts.index,subsets[subsets.keys()[i]]) #i/3(row) and i%3(column) are the position of each subplot
        sub[i/3,i%3].set_title('ts'+str(i))
        sub[i/3,i%3].set_xlabel('Nonometers')
        sub[i/3,i%3].set_ylabel('Counts')
```
