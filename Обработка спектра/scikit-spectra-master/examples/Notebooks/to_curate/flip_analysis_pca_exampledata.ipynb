## Flipping function of a Series: flippy_s, making the plot monotonically decrease or increase.

```python
def flippy_s(s,style='decrease'):
    y_flippy_s = [s[0]]
    shift = 0
    for i in range(1,len(s)):
        dy = s[i]-s[i-1]
        
        if style == 'decrease':
            if dy > 0: 
               shift -= 2*dy

        
        elif style == 'increase':
            if dy < 0: 
               shift -= 2*dy
            
        else:
            raise AttributeError('style should be decrease and increase')
            
        y_flippy_s.append(s[i]+shift)         
    return y_flippy_s
```

## Apply function flippy_s on the average area of raw spectrum

```python
plt.plot(ls.index,flippy_s(Series(ls.icol(0))))
```

## Flippy function of a dataframe: flippy_df.

```python
ts_1=ts.apply(flippy_s,axis=1)
```
