# Sampling and Selecting Data

### Objective

This guide will go over some basics for sampling and selecting data from skspec `Spectra` objects.  `skspec` relies on a subset of [pandas indexing and selection](http://pandas.pydata.org/pandas-docs/stable/indexing.html).  In particular, Multiindex/heirarchal slicing is not supported in `skspec`.  `skspec` also introduces approximate slicing, as it's a common need in spectroscopy.

### Basic Column and Row selection

#### Step 1: Load two test datasets

We load two sets of test data; one with `nm` spectral units and one with `cm-1` units, which are ordered traditionally from large to small.  We will show slicing on both objects:

`skspec` follows pandas directly in terms of columns and row seleciton.  By default, a Spectra will index by columns.  Let's print out and index by the first 3 columns:

```python
from skspec.data import aunps_glass, solvent_evap

t1 = aunps_glass().as_varunit('m')
t2 = solvent_evap().as_varunit('m')

# Plot them side by side
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(8,5))
t1.plot(ax=ax1)
t2.plot(ax = ax2);
```

```python
first_3 = t1.columns[0:3]
print first_3
t1[first_3]
```

For more on the various ways to select data from dataframes, see the [[pandas indexing and selection](http://pandas.pydata.org/pandas-docs/stable/indexing.html) tutorial.  We will focus mainly on slicing below, which is where `skspec` implements its own slicing objects.

### Approximate slicing

We can use `ts.nearby` to do approximate index slicing.  

-  By default, this work on the spectral/row axis. 
- **The limits provided must be within the limits of the dataset**.  
- This is inherently label based! 
    - Meaning even integers will be interpreted as wavelengths. 

```python
t1.nearby[550.0:650.0].plot(cbar=True);
```

We can slice t2, which has monotonically *decreasing* values, merely by slicing from large to small:

```python
#Integers still interpreted as labels, not index
t2.nearby[3000:2800].plot(cbar=True, colormap='cool');
```

#### Slicing both rows and columns

Pandas indexers natively support row and column slicing.  Because our data is in minutes, it's easy to slice.  This gets a little more tricky when data is in a complex unit like datetimeindex (see subsequent sections).  The syntax for row and column slicing is:

   - ts.nearby[**rowstart:rowend**, **colstart:colend**]

The columns slicing is also inherently label-based, meaning we cannot slice by index (eg ts.nearby[5:8, 9:1])

```python
t2.nearby[3000.0:2800.0, 3.0:11.0].plot(cbar=True, colormap='cool');
```

#### Individual and row and columns

Since `skspec` is using pandas slicing, it is very flexible.  In addition to slices, we can use individual rows/columns, or a list of rows/columns.  For example, to 3 column:

```python
t2.nearby[3000.0:2800.0, [2, 5, 7]]
```

We can also use steps for slicing to sample (eg plot 1 out of ever 3 timepoints).  Let's plot all wavelengths under 500nm, and every 10 columns from the first dataset

```python
t1.nearby[:500, ::10].plot(title='Time sampling by 10', cbar=True);
```

### Pandas slicing (label, index and mixed types)

Pandas slicers are not approximate; they require exact column ranges or index values.  Pandas implements 3 slicing objects supported by `puvvis`:
 
 - `ts.loc[]`: **Exact Label slicing**
 - `ts.iloc[]`: **Index slicing**
 - `ts.ix[]`: **Mixed label and index slicing**

#### `loc`: exact label slicing

`loc` is very similiar to `nearby`, except labels must be exact, and slicing outside the range of values of the index is acceptable; it merely returns an empty container:

```python
print t1.index[50], t1.index[55]
```

```python
t1.loc[448.64:450.48].plot(title='Location slicing based on index values above', cbar=True);
```

```python
#Values are outside valid spectral axis
t1.loc[5000.0:6000.0]
```

#### `iloc`: index slicing

`iloc` will take indicies for row and column values.  All aformentioned slice types, including single index, slices, list of index values etc... are supported; this is just wrapping the DataFrame's `iloc` slicer.  skspec is adding nothing to this.

Let's slice the first 100 wavelengths, and select 5 timepoints at indicies *5, 20, 30, 50, 55*

```python
t1.iloc[0:100, [5,20,30, 50, 55]].plot(cbar=True);
```

#### `ix`, mixed slicing.

`ix` will mix labels and indicies.  This is especially useful for timestamped data, where it gets tedious to pass in labels.  For example, let's convert our data back to timestamps and look at the first 2 columns:

```python
t1_stamped = t1.as_varunit('dti') #dti = datetimeindex
t1_stamped.columns[0:2]
```

While slicing/selection of timestamps is possible, it's sometimes easier to deal with indicies.  In this case, we can use `ix` to select the spectral axis through its labels and the perturbation axis (time) through indicies.  **Recall, label selection still requires exact values**

```python
t1_stamped.ix[448.64:450.48, 50:60].plot(title='Columns 50-60, rows sliced by label', color='r');
```
