# Principle Component Analysis: Part 1

### Objective

We introduce the basic builtin PCA interface in skspec, and then show how more complete decomposition modules, for example PCA in scikit learn, are easily incorporated when desired.

`skspec` presents a very basic interface to PCA, with the most common tasks such as computing the components, returning the eigenvectors/eigenvalues, projecting new data onto the components, and measuring the variance distribution in each component.  Several other interest PCA features like whitening, kernel PCA and so on are not builtin, but we will show how it's very easy to implement them from other resources such as scikit learn.

**Much of this interface is altered from a subset of [Alexis Mignon's pypca modules](http://code.google.com/p/pypca/source/list), so thank you for making this available!**

### Builtin PCA

The PCA functionality is stored on the Corr2d class; this was a design choice since in our work, correlation analysis and PCA are usually done together.  There's really no reason that the PCA aspects couldn't have been independently, and we'll show how to integrate other PCA modules in the later sections.  For now, let's get some testdata into the `Corr2d` class.  Please refer to the [Correlation Analysis tutorial part 1](http://nbviewer.ipython.org/urls/raw.github.com/hugadams/skspec/a47ff3280704f45badbd98b73dd2baac187dce2e/examples/Notebooks/correlation_p1.ipynb) for more explantion of this class.

```python
from skspec.data import aunps_water
ts = aunps_water().as_iunit('a').ix[220.0:700.0, :] #Slice wavelengths 220-700
ts.plot(cbar=True);
```

#### Next, initialize the correlation spectroscopy interface and mean-center the data:

```python
from skspec import Corr2d
cspec = Corr2d.from_spectra(ts)
cspec.center()
cspec
```

The `pca_fit()` method will compute the princple components, and by default, projects the data onto these components.

```python
components = cspec.pca_fit(n_components=5)
print components.shape
plt.plot(components);
```

```python
from skspec.plotting import specplot
from pandas import DataFrame
from skspec import TimeSpectra

specplot(TimeSpectra(cspec.data))
```

```python
print cspec.PCA.W.shape
print cspec.PCA.U.shape
print cspec.PCA.S.shape
print cspec.PCA.VT.shape
```

### Scikit Learn PCA

Let's use PCA in scikit learn, and the singular value decomposition.  PCA will mean center.  Scikitlearn PCA **expects (samples x features) ** and centers along the sample dimension.  If we pass our data transposed, it will center along the feature dimension.  For a discussion on SVD vs. PCA as well as shape considerations, see:

http://sourceforge.net/p/scikit-learn/mailman/message/32678084/

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
x = pca.fit_transform(ts)
```

```python
plt.plot(x)
```

```python
def svd_t(X):
    #x.shape = n_features,n_samples
    #every sample is a column
    mean= np.mean(X,axis=1)
    mean = mean[:,np.newaxis]
    X = X -mean
    u,s,v = np.linalg.svd(X,full_matrices=False)
    return u,s,v

u,s,v = svd_t(np.asarray(ts))
```
