```python
points = list(range(1,15))
names = list(map(lambda x: 'point_{}'.format(x), points))
```

```python
data = pd.read_excel('ExcelData.xlsx', header=None, names = names)
```

```python
weights = [np.sin(2* np.pi * i) for i in range(14)]
```

```python
arr_sin = list(range(0,13))
str_sin = list(range(0,14))
full = list()


for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        str_sin[j] = np.mean((weights[j] * data.iloc[i][j]))
    full.append(sum(str_sin))   
```

```python
new = list(range(1000))
for i in range(len(full)):
    if i == 0:
        new[i] = full[i]
    elif i < 998:
        new[i] = (full[i] + full[i-1]) / 2
    else:
        new[i] = full[i]
```

```python
arr_sin = list(range(0,13))
str_sin = list(range(0,14))
full = list()


for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        str_sin[j] = np.sin((2 * np.pi * data.iloc[i][j]) / 14)
    full.append(sum(str_sin))    
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(0,1000), new[0:1000])
```


# LOWESS (Locally Weighted Scatterplot Smoothing)

```python
from statsmodels.nonparametric.smoothers_lowess import lowess
```

```python
filtered = lowess(new, range(0,1000), is_sorted=True, frac=0.025, it=0)

plt.figure(figsize=(20,10))
plt.plot(range(0,1000), new, 'r')
plt.plot(filtered[:,0], filtered[:,1], 'b')
```

What is Lowess?
https://www.statsmodels.org/devel/generated/statsmodels.nonparametric.smoothers_lowess.lowess.html
The simplest definition of Locally Weighted Scatterplot Smoothing (LOWESS) is that it is a method of regression analysis which creates a smooth line through a scatterplot. This line provides a means to figure out relationships between variables. At the same time this line helps us understand trends of variables.  This method has advantages and disadvantages.

# Фильтра Баттерворта

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt

def butter_lowpass(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return b, a

def butter_lowpass_filtfilt(data, cutoff, fs, order=5):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = filtfilt(b, a, data)
    return y


time = range(0,1000)
x = new
cutoff = 3100
fs = 50000
new_smooth = butter_lowpass_filtfilt(new, cutoff, fs)


figure_pressure_trace = plt.figure(figsize=(20,10))
figure_pressure_trace.clf()
plt.plot(time, new, 'r', linewidth=1.0)
plt.plot(time, new_smooth, 'b', linewidth=1.0)
plt.show()
plt.close()
```

Второе предложение - использовать scipy.signal.filtfilt вместо lfilter для применения фильтра Баттерворта. filtfilt - фильтр с обратной связью. Он применяет фильтр дважды, один раз вперед и один раз назад, что приводит к задержке нулевой фазы.

Здесь представлена ​​измененная версия вашего script. Существенными изменениями являются использование filtfilt вместо lfilter и изменение cutoff от 3000 до 1500. Возможно, вам захочется поэкспериментировать с этим параметром - более высокие значения приводят к лучшему отслеживанию начала давления но слишком высокое значение не отфильтровывает 3 кГц (грубые) колебания после повышения давления.
